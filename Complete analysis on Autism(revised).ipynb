We are calculating the percentage of positive ASD for toddlers and adults whose QChat-10-Score is greater than 3

```python
sns.set_style('whitegrid')
data1= toddlers_df[toddlers_df['traits']=='Yes']
data2= adults_df[adults_df['traits']=='Yes']
print("Adults: ",len(data2)/len(adults_df) * 100)
print("Toddlers:",len(data1)/len(toddlers_df) * 100)
```

Around 1% of the population has ASD,but for this sample where we have taken the probable cases of Autism which are more then the 25 percentile of the datasets we see that we get 88% cases of toddlers with ASD(QChat Score >=3) whereas in case of 32% of adults (with AQ score >= 3 ) with positive ASD

```python
# we will comparing two datasets.
# Data: Born with jaundice based on Gender
fig, ax = plt.subplots(1,2,figsize=(20,6))
sns.countplot(x='Jaundice',data=data2,hue='Gender',ax=ax[0])
ax[0].set_title('ASD positive Adults born with jaundice based on gender')
ax[0].set_xlabel('Jaundice while birth')
sns.countplot(x='Jaundice',data=data1,hue='Gender',ax=ax[1])
ax[1].set_title('ASD positive Toddlers born with jaundice based on gender')
ax[1].set_xlabel('Jaundice while birth')
    
```

We can see here almost 6-7 times more(in Adults) and 2-3 times more(in Toddlers) of non-jaundice born ASD positive whereas according to reports that is around 10 times. Jaundice born children have a weak link with ASD.

According to reports, ASD is more common among boys than among girls. In case of Toddlers we see its nearly 3 times 
boys than girls. But in case of adults we see a contrary where the female is in higher ratio.

```python
#To visualize the distribution of autism in family within different ethnicity
fig, ax = plt.subplots(1,2,figsize=(20,6))
sns.countplot(x='Family_mem_with_ASD',data=data2,hue='Ethnicity',palette='rainbow',ax=ax[0])
ax[0].set_title('Positive ASD Adult relatives with Autism distribution for different ethnicities')
ax[0].set_xlabel('Adult Relatives with ASD')
sns.countplot(x='Family_mem_with_ASD',data=data1,hue='Ethnicity',palette='rainbow',ax=ax[1])
ax[1].set_title('Positive ASD Toddler relatives with Autism distribution for different ethnicities')
ax[1].set_xlabel('Toddler Relatives with ASD')
plt.tight_layout()
```

We can observe that both in Adults and Todddlers, White European ethnicities have a higher chance of being ASD positive if they have it in their genes. 

We cannot conclude anything firmly but we can stay confident that there is a genetic link for ASD positive as backed by studies
