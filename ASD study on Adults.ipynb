We will be dividing the ages in age-groups

```python
def compute_agegroup(adult_df):
    if(adult_df['age'] < 18):
        return 'Teens'
    
    elif(adult_df['age'] >= 18) & (adult_df['age'] <= 25):
        return 'Youth'
    elif(adult_df['age'] > 25) & (adult_df['age'] <= 35):
        return 'Adults(Young)'
    elif(adult_df['age'] > 35) & (adult_df['age'] <= 50):
        return 'Adults(Mid)'
    elif(adult_df['age'] > 50) & (adult_df['age'] <= 65):
        return 'Adults(Seniors)'
    
    elif(adult_df['age'] > 65) :
        return 'Seniors'
    
    
    
    
```

```python
adult_df['age-group'] = adult_df.apply(compute_agegroup, axis = 1)
```

To detect autism in adult one has to take the quiz and answer 10 questions. Based on A1-A10 questions one gets a AQ-10 Score. The result shows that it has highest correlation with A9,A6,A5 . So we can conclude that an individual who is unable to work out someone's feelings and thinkings, who is unable to understand whether someone listening to him is getting bored and someone who faces difficulty in reading between the lines when someone is talking to them probably have ASD traits.

In this data also we see a positive but a low correlation with AQ-Score/result. So we can infer that ASD can occur to an individual at any age.
