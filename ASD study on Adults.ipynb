```python
def compute_agegroup(adult_df):
    if(adult_df['age'] < 18):
        return 'Teens'
    
    elif(adult_df['age'] >= 18) & (adult_df['age'] <= 25):
        return 'Youth'
    elif(adult_df['age'] > 25) & (adult_df['age'] <= 35):
        return 'Adults(Young)'
    elif(adult_df['age'] > 35) & (adult_df['age'] <= 50):
        return 'Adults(Mid)'
    elif(adult_df['age'] > 50) & (adult_df['age'] <= 60):
        return 'Adults(Seniors)'
    
    elif(adult_df['age'] > 60) :
        return 'Seniors'
    
    
    
    
```

```python
adult_df['age-group'] = adult_df.apply(compute_agegroup, axis = 1)
```
